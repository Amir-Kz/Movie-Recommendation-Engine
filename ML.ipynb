## Feature Selection Part

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'overview', 'genre']]
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['genre']
```

```python
movies
```

```python
newData = movies.drop(columns = ['overview', 'genre'])
```

```python
newData
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 10000, stop_words = 'english')
```

```python
cv
```

```python
vector = cv.fit_transform(newData['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
newData[newData['title'] == "The Godfather"].index[0]
```

```python
sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse = True, key = lambda vector:vector[1])
for i in distance[0:5]:
    print(newData.iloc[i[0]].title)
```

```python
def recommand(movies):
    index = newData[newData['title'] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda vector:vector[1])
    for i in distance[0:5]:
        print(newData.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```
